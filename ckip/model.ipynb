### map 大以及小類 to integer
步驟一：計算大小累的長度<br>
步驟二：利用 dictionary map 到 integer<br>
步驟三：將大小類的資料以 integer 代替<br>

```python
sample_list = [df]
```

```python
#計算大小類的長度
small = len(df['小類'].unique())
big = len(df['大類'].unique())

class my_dictionary(dict):
 
  # __init__ function
  def __init__(self):
    self = dict()
 
  # Function to add key:value
  def add(self, key, value):
    self[key] = value

dict_small_obj = my_dictionary()
dict_big_obj = my_dictionary()

for i in range(big): #key(大類名稱):value（index）
    dict_big_obj.add(df['大類'].unique()[i], "%02d" %i)     

for i in range(small): #key(小類名稱):value（index）
    dict_small_obj.add(df['小類'].unique()[i], "%03d" %i)

print(dict_big_obj)
print(dict_small_obj)
for dataset in sample_list: #大小類更換成數值
    dataset['小類'] = dataset['小類'].map(dict_small_obj)
    dataset['大類'] = dataset['大類'].map(dict_big_obj)
```

```python
df['category_num'] = df['大類']+' '+df['小類'] # 合併大小類為一個 catogpry_num
df
```

### model 建立
步驟一：將資料向量化<br>
步驟二：測試模型準確率<br>

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
```

```python
# 將輸入的斷詞結果轉為轉為list
ckip_list = df['ckip_results'].tolist()
for c in range(len(ckip_list)):
  ckip_list[c] = ckip_list[c].replace('[[', '')
  ckip_list[c] = ckip_list[c].replace(']]', '')
  ckip_list[c] = ckip_list[c].replace('\'', '')
  ckip_list[c] = ckip_list[c].split(', ')
```

```python
vectorizer = CountVectorizer()

vocab = []

for l in range(len(ckip_list)):
  for v in range(len(ckip_list[l])):
      vocab.append(ckip_list[l][v])
```

```python
# 向量化
vectorizer.fit(vocab)
```

#### 訓練不同類別（大、小、總）目標
* 確認預測方法的準確率


* 利用大類訓練模型
* 確認大類的準確率

```python
X = df['ckip_string'].values.astype('U')
y = df['大類'].values
X = vectorizer.transform(X)
```

```python
# 切分訓練集與測試集，測試集占資料的30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# 使用多項貝氏分類器訓練模型
model = MultinomialNB()
model.fit(X_train, y_train)
predicted = model.predict(X_test)
print(model.score(X_test, y_test))
```
